```python
import matplotlib.pyplot as plt

```

```python
# Sample data
x = [0, 6, 14, 22, 28, 32, 38]
y1 = [15, 15, 12, 5, -1, -5, -10]
y2 = [0, 0, -20, -40, -50, -60, -70]

# Create a scatter plot
plt.scatter(x, y1, color='blue', label='150 MHz (main peak)')
plt.scatter(x, y2, color='red', label='300 MHz (second harmonic)')

# Add titles and labels
plt.title('Power of vs attenuation for 1st and 2nd harmonic', fontsize = 14)
plt.xlabel('attenuation set on QCM RF (a.u.)', fontsize = 14)
plt.ylabel('Power (dBm)', fontsize = 14)

# Increase font size of x and y ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add a legend
plt.legend(fontsize = 14)

# Add grid
plt.grid(True)
# Show the plot
plt.show()
```
